```python
at = []
bt = []
pro = []
prio = []

n = int(input("Enter the number of process: "))

for i in range(1, n+1):

  pro.append("P%d" %i)
  a = int(input("P%d arrival time: " %i))
  at.append(a)
  b = int(input("P%d burst time: " %i))
  bt.append(b)
  c = int(input("P%d priority: "%i))
  prio.append(c)


import pandas as pd
chart = {"Process No" : pro,
         "Arrival Time" : at,
         "Burst Time" : bt,
         "Priority" : prio}

df = pd.DataFrame(chart)
print()
print(df)


df = df.sort_values(by='Priority')
print()
print("Sorted Data: ")
print()
print(df)


new_pro = list(df["Process No"])
new_at = list(df["Arrival Time"])
new_bt = list(df["Burst Time"])

print()
print("Process No: ")
print(new_pro)
print("Arrival Time: ")
print(new_at)
print("Burst Time: ")
print(new_bt)


ct = []
calc = 0

for i in range(n):
  if i == 0:
    if new_at[i] == 0:
      calc = calc + new_bt[i]
      ct.append(calc)
    elif new_at[i] != 0:
      calc = calc + new_at[i] + new_bt[i]
      ct.append(calc)
  elif i > 0:
    if new_at[i] < calc:
      calc = calc + new_bt[i]
      ct.append(calc)
    else:
      d = new_at[i] - calc
      calc = calc + new_bt[i] + d
      ct.append(calc)

print()
print("Process No: ")
print(new_pro)
print("Completion time: ")
print(ct)


tat = []
wt =[]

for i in range(n):
  tat.append(ct[i] - new_at[i])
  wt.append(tat[i] - new_bt[i])

print()
print("Turn around time: ")
print(tat)
print()
print("Waiting time: ")
print(wt)


chart = {"Process No" : new_pro,
         "Arrival Time" : new_at,
         "Burst Time" : new_bt,
         "Completion Time" : ct,
         "Turn Around Time" : tat,
         "Waiting Time" : wt}
df2 = pd.DataFrame(chart)
df2 = df2.sort_values(by=['Process No'])

print()
print("Average Waiting Time: ", df2["Waiting Time"].mean())
print()
print("Average Turn Around Time: ", df2["Turn Around Time"].mean())

print()
df2
```
