```python
#updated SRTN
at = []
bt = []
pro = []
info = []

n = int(input("Enter the number of process: "))

for i in range(1, n+1):

  pro.append("P%d" %i)
  a = int(input("P%d arrival time: " %i))
  at.append(a)
  b = int(input("P%d burst time: " %i))
  bt.append(b)
  info.append([bt[i-1], at[i-1], i-1])
print(info)


tracker = []

for i in range(sum(bt)):
  
  l = [j for j in info  if j[1] <= i]
  l.sort(key=lambda x : x[0])
  info[info.index(l[0])][0] -= 1
  for k in info:
    if k[0] == 0:
      print("k is : ", k)
      info.pop(info.index(k))
      tracker.append([k, i+1])

print(tracker)
ct = [0] * (n)

for i in tracker:
  ct[i[0][2]] = i[1]
print(ct)


import pandas as pd
chart = {"Process No" : pro,
         "Arrival Time" : at,
         "Burst Time" : bt,
         "Completion Time" : ct}
df = pd.DataFrame(chart)

df["Turn Around Time"] = df["Completion Time"] - df["Arrival Time"]
df["Waiting Time"] = df["Turn Around Time"] - df["Burst Time"]


print()
print("Average Waiting Time: ", df["Waiting Time"].mean())
print()
print("Average Turn Around Time: ", df["Turn Around Time"].mean())

print()

df
```
