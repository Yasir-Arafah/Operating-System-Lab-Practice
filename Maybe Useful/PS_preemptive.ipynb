```python
#ps pre emptive
at = []
bt = []
pro = []
info = []
prio = []

n = int(input("Enter the number of process: "))

for i in range(1, n+1):

  pro.append("P%d" %i)
  a = int(input("P%d arrival time: " %i))
  at.append(a)
  b = int(input("P%d burst time: " %i))
  bt.append(b)
  c = int(input("P%d priority: "%i))
  prio.append(c)
  info.append([bt[i-1], at[i-1], prio[i-1], i-1])


print("info is: ", info)



tracker = []
calc = 0
at_tracker = at
at_tracker.sort()

if min(at) != 0:
  calc = calc + min(at)
else:
  calc = calc

new_calc = 0

for i in range(calc, len(at)):

  print("inside for loop i is: ", i)
  calc = calc = at_tracker[i]
  print("when i is %d calc is: "%i, calc)
  try:
    nat = at_tracker[i+1]
  except IndexError:
    nat = 'null'
  print("when i is %d nat is: "%i, nat)
  if nat == 'null':
    nat = 0
  else:
    nat = nat
  print("after if else nat is: ", nat) 
  l = [j for j in info  if j[1] <= calc]
  print("when i is %d calculating with calc and l is: "%i, l)
  l.sort(key=lambda x : x[2])
  print("sorted l is: ", l)
  if nat == 0:
    count = l[0][0]
  else:
    count = nat - calc
  print("when i is %d count is: "%i, count)
  info[info.index(l[0])][0] -= count
  print("updated info is: ", info)
  for k in info:
    if k[0] == 0:
      print("k is : ", k)
      info.pop(info.index(k))
      new_calc = calc + count
      tracker.append([k, new_calc])
    

print("tracker is: ", tracker)
print("updated info is: ", info)  
print("new_calc is: ", new_calc)


print("new for loop start")
for i in range(0, len(info)):
  l = [j for j in info  if j[1] <= new_calc]
  print("when i is %d calculating with calc and l is: "%i, l)
  l.sort(key=lambda x : x[2])
  print("sorted l is: ", l)
  count = l[0][0]
  print("when i is %d count is: "%i, count)
  info[info.index(l[0])][0] -= count
  print("updated info is: ", info)
  for k in info:
    if k[0] == 0:
      print("k is : ", k)
      info.pop(info.index(k))
      new_calc = new_calc + count
      print("after popping new calc is: ", new_calc)
      tracker.append([k, new_calc])


print("tracker is: ", tracker)
print("updated info is: ", info) 


ct = [0] * (n)

for i in tracker:
  ct[i[0][3]] = i[1]
print("ct is: ", ct)


import pandas as pd
chart = {"Process No" : pro,
         "Arrival Time" : at,
         "Burst Time" : bt,
         "Priority" : prio,
         "Completion Time" : ct}
df = pd.DataFrame(chart)

df["Turn Around Time"] = df["Completion Time"] - df["Arrival Time"]
df["Waiting Time"] = df["Turn Around Time"] - df["Burst Time"]


print()
print("Average Waiting Time: ", df["Waiting Time"].mean())
print()
print("Average Turn Around Time: ", df["Turn Around Time"].mean())

print()

df2 = df.loc[:,'Process No':'Completion Time']
df2
```
